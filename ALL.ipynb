```python
df.dropna(axis=1, inplace=False) #0 for row, 1 for col.     inplace means that df will be updated
```

```python
df.dropna(axis=0,thresh=len(df.columns)) #if thresh is 1 then if there is 1 non NA value then it will not be dropped
```

```python
df.replace(['?','<','NAN'],[np.nan,np.nan,np.nan], inplace=True)
df
```

```python

df["Pregnancies"] = pd.to_numeric(df["Pregnancies"], downcast="float")
df["BloodPressure"] = pd.to_numeric(df["BloodPressure"], downcast="float")
df["SkinThickness"] = pd.to_numeric(df["SkinThickness"], downcast="float")
df["Insulin"] = pd.to_numeric(df["Insulin"], downcast="float")
df["BMI"] = pd.to_numeric(df["BMI"], downcast="float")
df["DiabetesPedigreeFunction"] = pd.to_numeric(df["DiabetesPedigreeFunction"], downcast="float")
df["Age"] = pd.to_numeric(df["Age"], downcast="float")
df["Actual"] = pd.to_numeric(df["Actual"], downcast="float")
df["Pregnancies"].fillna(df['Pregnancies'].median(), inplace=True)
df['BloodPressure'].fillna(df['BloodPressure'].median(),inplace=True)
df['SkinThickness'].fillna(df['SkinThickness'].median(),inplace=True)
df['Insulin'].fillna(df['Insulin'].median(),inplace=True)
df['BMI'].fillna(df['BMI'].median(),inplace=True)
df['DiabetesPedigreeFunction'].fillna(df['DiabetesPedigreeFunction'].median(),inplace=True)
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Actual'].fillna(df['Actual'].median(),inplace=True)
df
```

```python
from sklearn.metrics import confusion_matrix
y_true = df['Actual']
y_pred = df['Predicted']
cm = confusion_matrix(y_true, y_pred)
print(cm)
```
